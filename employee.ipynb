```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
```

```python
# loading the dataset
df = pd.read_csv(r'Employee dataset.csv')
```

```python
#dropping the id column
df.drop('employee_id',axis = 1,inplace = True)
```

```python
# Accessing first 5 rows
df.head()
```

```python
#accessing last 5 rows
df.tail()
```

```python
# finding the null values
df.isna().sum()
```

```python
df.columns
```

```python
# filling the null values

df['previous_year_rating'].fillna( df['previous_year_rating'].median(), inplace=True)
```

```python
df['education'].fillna( df['education'].mode()[0], inplace=True)
```

```python
df.isna().sum()
```

```python
#finding the datatypes
df.dtypes
```

```python
# using the labelEncoder to convert into same datatype
le = LabelEncoder()
for i in df.columns:
    if df[i].dtype == 'object':
        df[i] = le.fit_transform(df[i])
print(df)
```

```python
df.dtypes
```

```python
# splitting the data
x = df.drop('is_promoted',axis = 1)
y = df['is_promoted']
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.3,random_state=70)
```

```python
x_train
```

```python
x_test
```

```python
y_train
```

```python
y_test
```

```python
from sklearn.naive_bayes import GaussianNB
g=GaussianNB()
g.fit(x_train,y_train)
y_pred = g.predict(x_test)
acc = accuracy_score(y_test,y_pred)
print(acc)
```

```python
confusion_matrix(y_test,y_pred)
```

```python
print(classification_report(y_test,y_pred))
```

```python
from xgboost import XGBClassifier
xg  = XGBClassifier()
xg.fit(x_train,y_train)
```

```python
y_pred = xg.predict(x_test)
acc = accuracy_score(y_test,y_pred)
print(acc)
```

```python
confusion_matrix(y_test,y_pred)
```

```python
print(classification_report(y_test,y_pred))
```

```python
from sklearn.svm import SVC
s = SVC()
s.fit(x_train,y_train)
```

```python
y_pred = s.predict(x_test)
acc = accuracy_score(y_test,y_pred)
print(acc)
```

```python
confusion_matrix(y_test,y_pred)
```

```python
print(classification_report(y_test,y_pred))
```

```python
from sklearn.ensemble import VotingClassifier
naive_bayes_classifier = GaussianNB()
svc_classifier = SVC(probability=True)
v = VotingClassifier(estimators=[
                ('naive_bayes', naive_bayes_classifier),
                ('svc', svc_classifier)
            ], voting='soft')
     
```

```python
v.fit(x_train,y_train)
```

```python
y_pred = v.predict(x_test)
acc = accuracy_score(y_test,y_pred)
print(acc)
```

```python
# prediction
abc = [[7,	11,	0	,0	,2,	1,	49,	4.0,	17	,0,	0,	47]]
a = v.predict(abc)
print(a)
```
